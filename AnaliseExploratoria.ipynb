#**Análise exploratória completa dos dados do ENEM de 2019**

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Tecnologia/Data Science/Projeto ENEM /Dataset_ENEM_2019.csv', encoding = 'ISO-8859-1', delimiter=',')
```

```python
df
```

```python
# Coletando todos os dados da maior nota média do ENEM 2019

indiceMaiorNotaMedia = df['NOTA_MEDIA'].idxmax() # Encontra o índice da linha que possui a maior nota_media
linhaMaiorNota = df.loc[indiceMaiorNotaMedia]    # Localiza a linha da maior nota e imprime todas as informações

display(linhaMaiorNota)
```

```python
# TIPO DE ESCOLA E NOTA MÉDIA

# Calculando a média das notas gerais das escolas públicas e privadas

tabelaTipoEscola = df.groupby('TP_ESCOLA')

tabelaMediaTipoEscola = tabelaTipoEscola[['NOTA_MEDIA']].mean()
tabelaMediaTipoEscola = tabelaMediaTipoEscola.reset_index()
display(tabelaMediaTipoEscola)

# Comparando nota média com o tipo de escola

plt.bar(tabelaMediaTipoEscola['TP_ESCOLA'], tabelaMediaTipoEscola['NOTA_MEDIA'], alpha=0.7)
plt.title('Relação entre nota média no ENEM 2019 e o tipo de escola')
plt.xlabel('Tipo escola')
plt.ylabel('Nota média')
plt.show()
```

```python
# COR/RAÇA E NOTA MÉDIA

tabelaCorRaca = df.groupby('COR_RACA')

tabelaMediaCorRaca = tabelaCorRaca[['NOTA_MEDIA']].mean()
tabelaMediaCorRaca = tabelaMediaCorRaca.reset_index()
display(tabelaMediaCorRaca)

# Comparando nota média com cor/raça

plt.scatter(tabelaMediaCorRaca['COR_RACA'], tabelaMediaCorRaca['NOTA_MEDIA'], alpha=0.7)
plt.title('Relação entre nota média no ENEM 2019 e cor/raça')
plt.xlabel('Cor/raça')
plt.ylabel('Nota média')
plt.show()
```

**Legenda:** 0 - Não declarado
1 - Branca
2 - Preta
3 - Parda
4 - Amarela
5 - Indígena

```python
# ESTADO DE RESIDÊNCIA E NOTA MÉDIA

# Calculando a média da nota geral de cada estado

tabelaEstados = df.groupby('UF_ESC')                      # agrupa todos os estados

tabelaMediaEstados = tabelaEstados[['NOTA_MEDIA']].mean() # cria uma tabela com os estados e a média de cada um deles
tabelaMediaEstados = tabelaMediaEstados.reset_index()
display(tabelaMediaEstados)

# Comparando nota média com o estado dos estudantes

plt.bar(tabelaMediaEstados['UF_ESC'], tabelaMediaEstados['NOTA_MEDIA'], width=0.5 ,alpha=0.7)
plt.title('Nota média no ENEM 2019 referente a cada estado')
plt.ylabel('Nota média')
plt.xlabel('Sigla dos estados')
plt.xticks(rotation=55, ha='right')
plt.show()
```

```python
# RENDA MENSAL E NOTA MÉDIA

# Calculando nota média de cada renda mensal

tabelaRendaMensal = df.groupby('RENDA_MENSAL')

tabelaMediaRendaMensal = tabelaRendaMensal[['NOTA_MEDIA']].mean()
tabelaMediaRendaMensal = tabelaMediaRendaMensal.reset_index()
display(tabelaMediaRendaMensal)

# Comparando nota média com a renda mensal

plt.scatter(tabelaMediaRendaMensal['RENDA_MENSAL'], tabelaMediaRendaMensal['NOTA_MEDIA'], alpha=0.7)
plt.title('Relação entre a nota média no ENEM 2019 e a renda mensal')
plt.ylabel('Nota média')
plt.xlabel('Renda mensal')
plt.show()
```
